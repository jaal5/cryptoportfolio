```python
lykke_trades['date'] = 0
lykke_trades['usd_equivalent'] = 0
lykke_trades['price'] = 0
for i, row in lykke_trades[lykke_trades['Type'] == 'Trade'].iterrows():
    curr_date = datetime.datetime(int(row.Date[6:10]),int(row.Date[0:2]),int(row.Date[3:5])).date()
    coin = row['Base Currency']
    
    lykke_trades.loc[i,'date'] = curr_date
    
    price = cryptocompare.get_historical_price(coin = coin, currency = 'USD', timestamp = pd.Timestamp(curr_date))[coin]['USD']
    lykke_trades.loc[i,'price'] = price
    lykke_trades.loc[i,'usd_equivalent'] = row['Base Amount']*price
    if i > 180:
        break
```
